Compile desired data into dataframe

```python
compiled = pd.DataFrame({'continent': ['Africa', 'Asia', 'Europe', 'North America', 'Oceana', 'South America', 'Rest of the World'],
                        '2017': list2017, '2018': list2018, '2019': list2019, '2020': list2020, '2021': list2021})
```

Construct and display data vis

```python
# import required libraries
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# set neon color scheme
colors3 = ['#08F7FE', '#FE53BB', '#F5D300']

# compile data and labels
x = compiled.drop('continent', axis=1).columns
y1 = compiled.iloc[6,:].to_list()[1:]
y2 = compiled.iloc[2,:].to_list()[1:]
y3 = compiled.iloc[3,:].to_list()[1:]
labels = ['North America', 'Europe', 'Rest of the World']

# smooth curves with cubic splines
cs1 = CubicSpline(x, y1)
cs2 = CubicSpline(x, y2)
cs3 = CubicSpline(x, y3)
xs = np.arange(2017, 2021, 0.00001)

# set up figure
fig, ax = plt.subplots(figsize=(11,7))
# plot stackplot
ax.stackplot(xs, cs1(xs), cs2(xs), cs3(xs), labels=labels, colors=colors3, baseline='sym', alpha=1)
# add in glow effect
for y in range(1, 15):
    cs1 = CubicSpline(x, [x + (y * 10) for x in y1])
    cs3 = CubicSpline(x, [x + (y * 10) for x in y3])
    ax.stackplot(xs, cs1(xs), cs2(xs), cs3(xs), colors=colors3, baseline='sym', alpha=0.03)
# tidy axes and labels to preference
ax.legend(reversed(plt.legend().legendHandles), labels, loc='upper left', labelcolor='linecolor',
         fontsize=14, handlelength=0)
ax.set_facecolor('#27344D')
fig.patch.set_facecolor('#27344D')
plt.xticks([2017.0, 2018.0, 2019.0, 2020.0, 2021.0], [2017, 2018, 2019, 2020, 2021],
           fontsize=20, color='0.9', weight='bold')
plt.yticks(fontsize=0)
plt.title('Relative Number of Survey Responses by Region', fontsize=20, color='0.9', loc='left',
         weight='bold')
plt.savefig('rainbow_road.png', dpi=300, facecolor='#27344D', edgecolor='#27344D')
plt.show;
```
